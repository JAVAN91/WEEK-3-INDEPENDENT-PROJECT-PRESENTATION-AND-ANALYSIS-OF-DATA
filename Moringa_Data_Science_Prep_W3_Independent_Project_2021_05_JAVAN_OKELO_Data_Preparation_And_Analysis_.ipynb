DATA ANALYSIS

Value Column contains the billing counts for the Telecom for the 3days

```python
#Value Column contains the billing counts for the Telecom for the 3days
New_df['VALUE'].describe()
```

Showing the value of each products in Telecom for the 3days

```python
#Showing the value of each products in Telecom for the 3days
pivot=pd.pivot_table(New_df,values=['VALUE'],index=['PRODUCT'],aggfunc='sum')
pivot.sort_values('VALUE',ascending=False)
```

Each city and the rates at which they use the products

CITIES/VILLES THAT WERE USED MOST IN THE THREE DAYS
